```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
csv_file = open('renthop11.csv', 'w', encoding='utf-8', newline='')
writer = csv.writer(csv_file)

for i in range(1,2485):
    print("Scraping page number " + str(i))
    web = 'https://www.renthop.com/search/nyc?max_price=50000&min_price=0&sort=hopscore&q=&search=0&page=' + str(i)
    response = requests.get(web)
    text = BeautifulSoup(response.text, 'html.parser')
    titleTags = text.find_all('div', {'class': "search-info pr-3 pl-3 pr-md-0 pl-md-4"})    

    for j in range(20):
        review_dict = {}
        #print("Scraping house number " + str(j))
        link = titleTags[j].find('a', {'class': "font-size-11 listing-title-link b"}).attrs['href']
        address = titleTags[j].find('a', {'class': "font-size-11 listing-title-link b"}).string
        borough = titleTags[j].find('div', {'class': "font-size-9 overflow-ellipsis"}).string
        score = titleTags[j].find('span', {'class': "b font-blue"}).string
        price = titleTags[j].find_all('td', {'class': "font-size-11 b"})[0].string
        bed = titleTags[j].find_all('td', {'class': "font-size-11 b"})[1].find('span').string
        bath = titleTags[j].find_all('td', {'class': "font-size-11 b"})[2].find('span').string
        amenity = titleTags[j].find_all('div', {'style': "margin-top: 8px;"})[1].text
    
        review_dict['link'] = link
        review_dict['address'] = address
        review_dict['borough'] = borough
        review_dict['score'] = score
        review_dict['price'] = price
        review_dict['bed'] = bed
        review_dict['bath'] = bath
        review_dict['amenity'] = amenity
        writer.writerow(review_dict.values())   
```
